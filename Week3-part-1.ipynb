Reading the table from source - Wikipedia URL to List of Postal Codes of Canada

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
read_table = pd.read_html(url,header=[0])
df = read_table[0]
df.head(10)
```

```python
#Renaming column headers
df = df.rename(index=str, columns={'Postcode':'PostalCode','Neighbourhood':'Neighborhood'})
df.head()
```

Ignore cells with a borough that is "Not assigned"

```python
df = df[df.Borough !='Not assigned']
df.reset_index(drop=True,inplace=True)
df.head()
```

Group neighborhoods in the same postal code area

```python
df = df.groupby('Postal Code',as_index=False).agg(lambda x: ','.join(set(x.dropna())))
df.head()
```

IF Neighborhood="Not assigned", then the value of the neighboorhood = the value the respective Borough

```python
df.loc[df['Neighborhood'] == 'Not assigned','Neighborhood'] = df['Borough']
df.head()
```
